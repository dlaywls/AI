```python
import numpy as np
import torch

import torch.nn as nn

import torch.optim as optim
import torch.nn.init as init
import matplotlib.pyplot as plt
```

```python
num_data=20
num_epoch=500

x =init. torch.Tensor(
    [155, 160, 165, 170, 175,
     180, 185, 190, 195, 200,
     176, 182, 148, 182, 178,
     150, 192, 144, 188, 167,
     148, 154, 164, 187, 164,
     146, 179, 152, 196, 199])

y = init.torch.Tensor(
    [51, 60, 60, 60, 68,
     68, 80, 80, 81, 86,
     67, 81, 40, 75, 72,
     45, 84, 38, 75, 53,
     40, 46, 51, 77, 57,
     39, 70, 44, 78, 90])




# 입력 특성과 목표값을 정규화
x_mean = x.mean()
x_std = x.std()
y_mean = y.mean()
y_std = y.std()

x_normalized = (x - x_mean) / x_std
y_normalized = (y - y_mean) / y_std

```

```python
plt.figure(figsize=(10,10))
plt.scatter(x.numpy(),y.numpy(),s=7,c="gray")

plt.show()
```

```python
model=nn.Linear(1,1)
loss_func=torch.nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=0.1)
```

```python

label=y_normalized.reshape(-1,1)
for i in range(50):
  optimizer.zero_grad()
  output=model(x_normalized.reshape(-1,1))
  loss=loss_func(output,label)
  loss.backward()
  optimizer.step()



```
